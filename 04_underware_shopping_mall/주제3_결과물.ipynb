# 가슴 사이즈 및 구독 회차별 유저 리텐션 계산.


### 가공
1. size를 입력하지 않은 데이터 제거한다.
2. 제대로 입력된 size 입력값만 골라서 list에다 담아둔다.
3. size에 제대로 입력된 size가 아닌 이상한 값이 들어있는 데이터를 지운다.
4. 중복된 id를 제거하기 위해 구독자를 맨 상위로 올린다.(구독이 적혀있는 것은 확실히 구독한 것이기 때문)
5. 지불날짜 가장 과거일자부터 정렬한다.
6. 사람의 수를 구하는 것으로 1회차 구독한 id를 제외한 중복된 id 제거한다.
7. list에 담아둔 size 항목으로 그룹으로 묶에서 가슴사이즈 입력한 사람의 수 구한다.(리텐션의 분모로 사용)

8. 1회차 구독한 수 사람의 수를 구한다.
9. 1회차 구독자 수를 list에 담아둔 size 항목으로 묶어서 사람의 수를 구한다.(리텐션의 분자로 사용)

10. 위의 size 항목으로 묶은 가슴사이즈 입력한 사람의 수를 분모로 하고, size 항목으로 묶은 1회차 구독자 수를 분자로 하여 리텐션을 구한다.


### 자동화
1. 1회차 구독자 수를 구했으면 다음으로 2회차 구독자 수를 구하여 '2회차 구독자 수/1회차 구독자 수' 로 분자 부분만 계속 구해서 다음 회차 리텐션을 같은 방식으로 구할 수 있게 할 수 있다.
2. 가공에서 지불날짜를 가장 과거일자부터 정렬한 데이터(결과값이라 부르겠음)를 가져와 중복된 값 중에서 맨 처음 값을 출력한다.
3. 중복된 값 중에서 맨 처음 값을 결과값에서 제거한다.
4. 유저 ID로 중복제거를 한 후 첫번째 값을 추출한다.(2번째 구독자 수)
5. size 항목으로 그룹하여 2번째 구독자 수를 출력한다.

6. 중복된 값 중에서 맨 처음 값을 제거한 결과값을 결과값으로 사용하여 위의 과정을 다시 실행한다(3번째 구독자 수)
7. 위의 과정을 반복하고 중복값이 0이 나올때가 마지막 구독자 수 이므로 마지막으로 size 항목으로 그룹하고 마무리한다.

이상하지만 결과값을 보자면 총 2회차 구독까지 1500명이 진행했으며 4회차부터는 구독자가 없었다. 사이즈 별 리텐션은 아래 데이터프레임으로 출력했다.


### 해석
리텐션율은 가슴사이즈 별로 많이 차이나지 않아 느낌으로는 가슴사이즈 별로 우리의 서비스가 차별이 있는 것 같지 않다.
제일 좋은 것은 모두 서비스를 만족하는 것이지만 리텐션율을 보면 그런 것 같지는 않다. 오히려 최악인 모두 불만족일 수 있는 경우는 의심해 봐야 할 것 같다.
우리의 서비스가 가슴사이즈별로 정말 다른 만족도를 보이는지
그리고 가슴사이즈별 구독, 가슴사이즈별 구독 횟수가 상관관계가 있는지 그래프를 통해 산점도를 그려보거나 피어슨 상관계수를 구하여 상관분석을 하는 작업이 필요할 것 같다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt

import warnings
warnings.filterwarnings('ignore')
```

```python
total_df = pd.DataFrame()
payment_df = pd.read_csv('./payment_window.csv', encoding='CP949')
bra_df = pd.read_csv('./bra_size_window.csv', encoding='CP949')
user_df = pd.read_csv('./user_window.csv', encoding='CP949')
pay_bra_df = pd.merge(payment_df, bra_df, how='outer', on='user_idx')
total_df = pd.merge(pay_bra_df, user_df, on='user_idx')
drop_cols = ['Unnamed: 0_x', 'Unnamed: 0', 'Unnamed: 0_y']
total_df.drop(drop_cols, axis=1, inplace=True)
total_df.reset_index(drop=True, inplace=True)

total_df['paid_date'] = pd.to_datetime(total_df['paid_date'])
total_df['joined_date'] = pd.to_datetime(total_df['joined_date'])
total_df['created_date'] = pd.to_datetime(total_df['created_date'])
```

## 데이터 가공

```python
# size를 입력하지 않은 행 제거
total_notnull_df = total_df.dropna(subset=['size'])
total_notnull_df.shape

# 사이즈가 이상하게 들어 있는 데이터 제거
drop_row = ['3.2', '밑둘레가 더 큼', '85C/D/E/F/', '올리브', '알수없음 75/7','밑둘레가 더 큼 7',\
           '밑가슴둘레가 더 큼, ', '알수없음(0/0)', '80C/D/E/F/', '75D/E/F/G', '알수없음 0/0', '70D/E/F/G',\
           '알수없음 0/0', '70D/E/F/G', '알수없음 76/11', '그린', '65D/E/F/G', '알수없음(67/7)', '(70/70)',\
           '밑둘레가 더 큼 8', '알수없음 70/12', '(70/70)', '밑둘레가 더 큼 8', '알수없음 70/12',\
           '알수없음 73/10', '82/11', '밑가슴둘레가 더 큼']

total_notnull_drop_df = total_notnull_df[~total_notnull_df['size'].isin(drop_row)]

# 가슴 사이즈의 모든 항목 list에 담기
size_list = total_notnull_drop_df['size'].unique()
size_list.sort()
```

```python
# 중복된 id를 제거하기 위해 구독자를 맨 상위로 올리기(구독이 적혀있는 것은 확실히 구독한 것이기 때문)
total_notnull_drop_df = total_notnull_drop_df.sort_values(by='subscription_type', ascending=True)

# 지불날짜 가장 과거일자부터 출력되도록 하기
total_notnull_drop_df = total_notnull_drop_df.sort_values(by='paid_date', ascending=True)
total_notnull_drop_df = total_notnull_drop_df.reset_index(drop=True)

# 사람의 수를 구하는 것으로 1회차 구독한 id를 제외한 중복된 id 제거 
total_notnull_drop_df_1 = total_notnull_drop_df.drop_duplicates(subset=['user_idx'], keep='first')
print('가슴 사이즈 입력한 사람의 수 : ',total_notnull_drop_df_1.shape[0])

# size 항목을 그룹으로 묶에서 가슴사이즈 입력한 사람의 수 구하기
total_notnull_drop_df_1_size = total_notnull_drop_df_1.groupby(by='size')['size'].count()
total_notnull_drop_df_1_size

# 1회차 구독한 수
subscription_df = total_notnull_drop_df_1.dropna(subset=['subscription_type'])
print('가슴 사이즈 입력한 사람의 수 중의 1회차 구독한 사람의 수 : ',subscription_df.shape[0])

# 1회차 구독자 수 size로 묶어서 보기
subscription_size_df = subscription_df.groupby(by='size')['size'].count()

final_df = pd.DataFrame(index=size_list, columns=['사이즈입력수_1','구독자수_1','구독자수/사이즈입력수_1'])
final_df['사이즈입력수_1'] = total_notnull_drop_df_1_size
final_df['구독자수_1'] = subscription_size_df
final_df['구독자수/사이즈입력수_1'] = final_df['구독자수_1']/final_df['사이즈입력수_1']
```

# 구독자 수 리텐션 자동화

```python
def retention_plus(total_df, final_df, subsc_num):
    result_df = pd.DataFrame()
    total_df = total_df.dropna(subset=['subscription_type'])
    
    for i in range(2,subsc_num):
        # 결과값에서 중복된 값 중에서 맨 처음 값을 출력한다.
        dupli_df = total_df[total_df.duplicated(subset=['user_idx'], keep='first')]
        print('{}번째 구독자 수 중복값 : '.format(i), dupli_df.shape[0])

        # 중복 중 맨 처음 값을 결과값에서 제거한다.
        total_df = total_df.drop(index=dupli_df.index)
        total_df = total_df.dropna(subset=['subscription_type'])

        # 중복제거를 한 후 첫번째 값을 추출한다.
        result_df = total_df.drop_duplicates(subset=['user_idx'], keep='first')

        # size로 그룹하여 숫자를 센다.
        result_size_df = result_df.groupby(by='size')['size'].count()
        
        # 사이즈 입력 수 구하기
        final_df['구독자수_'+str(i)] = result_size_df
        final_df['현재 구독자 수/이전 구독자 수_'+str(i)] = final_df['구독자수_'+str(i)]/\
                                                    final_df['구독자수_'+str(i-1)]

        
    return display(final_df)
```

```python
retention_plus(total_notnull_drop_df, final_df, 4)
```
